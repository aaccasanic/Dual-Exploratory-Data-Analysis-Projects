# EDA - Lifestyle and Sleep Patterns

[Lifestyle and Sleep Patterns Shortcut](https://www.kaggle.com/datasets/minahilfatima12328/lifestyle-and-sleep-patterns)

Dataset Overview

The dataset was extracted from Kaggle and contains 374 records with a total of 13 features related to lifestyle, sleep habits, and basic physiological measurements. The variables include:

- Gender

- Age

- Occupation

- Sleep Duration

- Sleep Quality

- Physical Activity Level

- BMI Category (Body Mass Index)

- Blood Pressure

- Heart Rate

- Daily Steps

- Sleep Disorder

This dataset provides a compact but insightful foundation for analyzing how lifestyle patterns—such as sleep behavior and physical activity—relate to individual health indicators.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df
```

Since Person ID is a unique identifier and does not carry analytical value, it will be excluded from all statistical computations.

To prevent it from being interpreted as a numeric field, the column will be explicitly cast to a string.

```python
df['Person ID'] = df['Person ID'].astype(str)
```

```python
df.info()
```

```python
df.shape
```

Missing values in the Sleep Disorder column represent individuals who do not report any sleep-related condition.

To ensure these records can be counted and visualized properly, the NaN values are replaced with the label "No Disorder".

```python
# Fill NaN values to ensure proper counting and visualization
Disorder = df['Sleep Disorder'].copy()
Disorder_fill = Disorder.fillna('No Disorder')
df['Sleep Disorder'] = Disorder_fill
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(10,16), facecolor='white', edgecolor='black')
plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.suptitle('Variable Distributions #1', fontsize=20, color='black', fontweight='bold')

plt.subplot(3,2,1)
sns.histplot(df['Gender'], bins=10, color='blue')
plt.title('Gender Distribution')

plt.subplot(3,2,2)
sns.histplot(df['Age'], kde=True, bins=10, color='blue')
plt.title('Age Distribution')

plt.subplot(3,2,3)
sns.histplot(df['Sleep Duration'], kde=True, bins=10, color='green')
plt.title('Sleep Duration Distribution')

plt.subplot(3,2,4)
sns.histplot(df['Quality of Sleep'], kde=True, bins=10, color='red')
plt.title('Sleep Quality Distribution')

plt.subplot(3,2,5)
sns.histplot(df['Physical Activity Level'], kde=True, bins=10, color='purple')
plt.title('Physical Activity Level Distribution')

plt.subplot(3,2,6)
sns.histplot(df['Stress Level'], kde=True, bins=10, color='orange')
plt.title('Stress Level Distribution')

plt.show()
```

```python
plt.figure(figsize=(10,16), facecolor='white', edgecolor='black')
plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.suptitle('Variable Distributions #2', fontsize=20, color='black', fontweight='bold')

plt.subplot(3,2,1)
sns.histplot(df['Heart Rate'], kde=True, bins=10, color='brown')
plt.title('Heart Rate Distribution')

plt.subplot(3,2,2)
sns.histplot(df['Daily Steps'], kde=True, bins=10, color='pink')
plt.title('Daily Steps Distribution')

plt.subplot(3,2,3)
sns.histplot(df['BMI Category'], kde=True, bins=10, color='gray')
plt.title('BMI Category Distribution')

plt.subplot(3,2,4)
sns.histplot(df['Blood Pressure'], kde=True, bins=10, color='cyan')
plt.title('Blood Pressure Distribution')

plt.subplot(3,2,5)
sns.histplot(df['Sleep Disorder'], kde=True, bins=10, color='magenta')
plt.title('Sleep Disorder Distribution')

plt.show()
```

## Correlational Analysis and Exploratory Data Analysis

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Numeric Feature Correlations')
plt.show()
```

```python
sns.boxplot(data=df, x="Gender", y="Sleep Duration", palette="Set2", hue='Gender',legend=False)
plt.title("Sleep Duration by Gender", fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(data=df, x="Physical Activity Level", y="Sleep Duration", scatter_kws={'alpha':0.4}, line_kws={'color':'red'})
plt.title("Physical Activity Level vs. Sleep Duration", fontweight="bold")
plt.show()
```

```python
df['Stress Level'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Stress Level", y="Quality of Sleep", palette="coolwarm", hue="Stress Level", legend=False)
plt.title("Stress Level vs. Sleep Quality", fontweight="bold")
plt.show()
```

```python
sns.violinplot(data=df, x="Sleep Disorder", y="Sleep Duration", palette="Set2", hue='Sleep Disorder', legend=False)
plt.title("Sleep Duration Distribution by Sleep Disorder", fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Occupation", y="Quality of Sleep", estimator=np.median)
plt.xticks(rotation=45, ha="right")
plt.title("Median Sleep Quality by Occupation", fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(14,5))
df['Age Group'] = pd.cut(df['Age'], bins=[0,25,35,45,55,65,100], labels=['<25','25-35','35-45','45-55','55-65','65+'])
sns.barplot(data=df, x='Age Group', y='Sleep Duration', estimator=np.mean, palette='viridis', hue= 'Age Group',  legend=False, dodge=False)
plt.title("Average Sleep Duration by Age Group", fontweight='bold')
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))
sns.histplot(df["Sleep Duration"], kde=True, ax=axes[0,0])
sns.boxplot(data=df, x="Gender", y="Sleep Duration", ax=axes[0,1])
sns.regplot(data=df, x="Physical Activity Level", y="Sleep Duration", ax=axes[1,0])
sns.boxplot(data=df, x="Stress Level", y="Quality of Sleep", ax=axes[1,1])
plt.tight_layout()

plt.subplots_adjust(top=0.9)
plt.suptitle("Sleep Analysis Dashboard", fontsize=20, fontweight="bold")
plt.show()
```

## Conclusions

- The average sleep duration is approximately 7.1 hours, which falls within the healthy range.

- There is a positive correlation between physical activity and sleep duration: individuals who engage in more exercise tend to sleep longer.

- Higher stress levels are associated with lower sleep quality, highlighting the importance of emotional and stress management.

- The variables most strongly related to sleep duration are sleep quality (0.88) and stress level (0.81).

- The data reveals a trend in which older individuals tend to have longer sleep durations.

- Overall, this analysis shows how combining behavioral and health metrics can generate valuable insights for promoting healthier and more balanced habits.
